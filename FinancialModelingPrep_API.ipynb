A 3-day window for certain quarters is lacking because the market is closed on holidays or weekends --> missing stock data on those days. So if that day is missing, window should jump to the next nerest trading dates --> **Trading-Day indexing**

```python
def get_event_window(df, event_day, window =3):
    df = df.sort_values("date").reset_index(drop=True) # avoid having t-3 as a future date, as the FMP returns stock data in descending order
    df["date"] = pd.to_datetime(df["date"])
    event_day  = pd.to_datetime(event_day)

    # If event day is not in trading date, move forward to the next trading day
    if event_day not in df["date"].values:
        event_day = df[df["date"] > event_day]["date"].iloc[0]

    idx = df.index[df["date"]== event_day][0] # [0] grabs the first element of that index list (the index list: Int64Index([***], dtype='int64'))
    return df.iloc[idx-window: idx+window+1] 
```

```python
Q3_2025 = get_event_window(df, "2025-07-31", 3)
Q2_2025 = get_event_window(df, "2025-05-01", 3)
Q1_2025 = get_event_window(df, "2025-01-30", 3)
Q4_2024 = get_event_window(df, "2024-10-31", 3)
```

## Define the north star metrics
1> Immediate retunr (t-1 to t+1) --> capture the impact of the earning itself on the market
2> Extended drift return (t-3 to t+3) --> check the trending of the initial reaction
3> Gap return (overnight reaction) --> capture the difference between event day 's open and previous day 's close

```python
def calculate_metrics(event_df):
    """
    event_df: DataFrame with 7 rows (t-3 ... t+3) for a quarter
    Assumes columns: 'date', 'open', 'close'
    """
    # t-3 and t+3
    t_minus3_close = event_df.iloc[0]['close']
    t_plus3_close = event_df.iloc[-1]['close']
    
    # t-1, t0, t+1
    t_minus1_close = event_df.iloc[3-1]['close']  # 4th row is t0, so t-1 is 3rd row
    t0_open = event_df.iloc[3]['open']
    t_plus1_close = event_df.iloc[3+1]['close']
    
    # Metrics
    immediate_return = (t_plus1_close - t_minus1_close) / t_minus1_close * 100
    extended_drift = (t_plus3_close - t_minus3_close) / t_minus3_close * 100
    gap_return = (t0_open - t_minus1_close) / t_minus1_close * 100
    
    return {
        'Immediate Return %': round(immediate_return, 2),
        'Extended Drift %': round(extended_drift, 2),
        'Gap Return %': round(gap_return, 2)
    }

# Apply to all quarters
metrics_Q4_2024 = calculate_metrics(Q4_2024)
metrics_Q1_2025 = calculate_metrics(Q1_2025)
metrics_Q2_2025 = calculate_metrics(Q2_2025)
metrics_Q3_2025 = calculate_metrics(Q3_2025)

# Combine into one DataFrame
df_metrics = pd.DataFrame([
    {'Quarter': 'Q4 2024', **metrics_Q4_2024},
    {'Quarter': 'Q1 2025', **metrics_Q1_2025},
    {'Quarter': 'Q2 2025', **metrics_Q2_2025},
    {'Quarter': 'Q3 2025', **metrics_Q3_2025},
])

print(df_metrics)
```
